```python
import os
from PIL import Image

# Set the path to your dataset folder
dataset_path = "TACO-master/data"

# Function to get image dimensions
def get_image_dimensions(folder_path):
    image_sizes = []
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.endswith((".jpg", ".jpeg", ".png")):  # Adjust extensions as needed
                file_path = os.path.join(root, file)
                try:
                    with Image.open(file_path) as img:
                        width, height = img.size
                        image_sizes.append((file_path, width, height))
                except Exception as e:
                    print(f"Error reading {file_path}: {e}")
    return image_sizes

# Get dimensions of all images
image_dimensions = get_image_dimensions(dataset_path)

# Print or save the dimensions
for file_path, width, height in image_dimensions:
    print(f"Image: {file_path}, Width: {width}, Height: {height}")

# Optionally, summarize the dataset
print(f"\nTotal images: {len(image_dimensions)}")
```
